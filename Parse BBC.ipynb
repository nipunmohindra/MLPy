```python
path = "/Users/rishushrivastava/Documents/AI/Dataset/bbc"

file_list = []
for dirpath,dirname,filename in os.walk(path):
    category = dirpath.split('/')[-1]
    [ file_list.append((category, dirpath+'/'+f)) for f in filename if '.txt' in f ]

file_list
```

```python
data_content = []

for category, files in file_list:
    with open(files,'r',encoding='latin1') as f:
        #print(f'reading file {files} in category: {category}')
        content = f.read()
        content_dict = {'category':category, 'content':content, 'filename':files}
        data_content.append(content_dict)
        
        
```

```python
data_content
```

```python
import json

output_path = '/Users/rishushrivastava/Documents/AI/Dataset/'

with open(output_path+'bbc_raw_dict.json','w') as of:
    json.dump(data_content, of)
```
