# All About Activation functions in Neural Network

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

import numpy as np
```

# Sigmoid Activation

```python
x = torch.range(-5.,5.,0.1)
y = torch.sigmoid(x)
```

```python
plt.plot(x.numpy(),y.numpy())
plt.grid(True)
plt.title('Sigmoid Activation')
plt.xlabel('x')
plt.xlabel('y')
plt.show()
```

# TanH Activation

```python
x = torch.range(-5.,5.,0.1)
y = torch.tanh(x)
```

```python
plt.plot(x.numpy(),y.numpy())
plt.grid(True)
plt.title('TanH Activation')
plt.show()
```

# ReLU Activation

```python
relu = nn.ReLU()

x = torch.range(-5.,5.,0.1)
y = relu(x)
```

```python
plt.plot(x,y)
plt.grid(True)
plt.title('ReLU Activation')
plt.show()
```

# Parametric ReLU (a variant of ReLU activation for dying gradient problem)

```python
prelu = nn.PReLU(num_parameters=1)

x = torch.range(-5.,5.,0.1)
y = prelu(x)
```

```python
plt.plot(x.detach().numpy(),y.detach().numpy())
plt.grid(True)
plt.title('PReLU Activation')
plt.show()
```

# Softmax

```python
softmax = nn.Softmax(dim=1)

x = torch.randn(1,3)
y = softmax(x)  ## generates the probability
```

```python
print(x)
print(y)
```
